## Exploratory Data Analysis

In the EDA step, I will visualize each feature by its class distribution. The "class" refers that the information of this mushroom is edible or poisonous.

### Mushroom class

```python
plt.figure(figsize=(6, 6), dpi=100)
plt.title("Distribution of the Mushrooms by their Classes")
plt.pie([df["class"].value_counts()[0], df["class"].value_counts()[1]], labels=["Edible", "Poisonous"], autopct='%1.1f%%', colors=["seagreen", "indianred"], wedgeprops = {"edgecolor" : "black", "linewidth": 2, "antialiased": True})
plt.legend()
plt.show()
```

As seen from the figure above, the red part represents that the percentage of the poisonous mushrooms in the dataset. The green part represents that the percentage of edible mushrooms. Since the percentages of these classes are pretty close to each other, we don't really need to balance the classes. From now on, I will represent the red color for the poisonous class and green for the edible class of the mushrooms.

### Cap Shapes

![mushroom-cap-shapes](images/mushroom-cap-shapes.webp)

```python
plt.figure(figsize=(12, 6), dpi=200)
plot = sns.countplot(x="cap-shape", data=df, hue="class", palette=["indianred", "seagreen"], order=df["cap-shape"].value_counts().index, edgecolor=(0, 0, 0), linewidth=2)
plot.set_xticklabels(["Convex", "Flat", "Knobbed", "Bell", "Sunken", "Conical"])
plt.legend(["Poisonous", "Edible"], loc="upper right")
plt.ylabel("Number of the Mushrooms", fontsize=12)
plt.xlabel("Types of the Cap Shapes of the Mushrooms", fontsize=12)
plt.title("Distribution of the Mushrooms by their Classes and Cap Shapes", fontsize=14)
```

According to this graph, the Convex Cap Shape is the most frequent. The Edible and Poisonous class mushrooms are pretty close to each other for the Convex and the Flat type. In contrast, for the following types which are Kobbed and Bell, the class types are pretty imbalanced distributed. The Sunken type is completely edible. The Conical type is completely poisonous.
